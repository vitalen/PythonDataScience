Получить матрицу B, умножив матрицу A на матрицу At с помощью матричного умножения. Какой размер имеет матрица B? Получится ли вычислить обратную матрицу для матрицы B и почему?

```python
a = np.array([i for i in range(0,12)])
print(a)
A = a.reshape(4,3)
print(A)

At = np.transpose(A)
print(At)

B = np.dot(A,At)
print(B)
print(B.size)



print('Если матрица квадратная, то мы можем посчитать её определитель. Oбратную матрицу можно посчитать только для матрицы, определитель которой не равен 0')
d = np.linalg.det(B)
print(f'определитель: {d}')
```
